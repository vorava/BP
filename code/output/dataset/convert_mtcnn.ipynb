```python
from mtcnn import MTCNN
import os
import cv2
import json

mtcnn_detector = MTCNN()
```

```python
def detect_faces(image):
    out_bboxes = []
    
    res = mtcnn_detector.detect_faces(image)
    for r in res:
        box = r["box"]
        out_bboxes.append([box[0], box[1], box[2], box[3]])

    
    nof_detections = len(res)
    
    return nof_detections, out_bboxes
```

```python
IMAGES_PATH = "../../api/data/workspace/aug/val/images/"
OUTPUT_PATH = f"results/MTCNN/result_MTCNN.json"

images = sorted(os.listdir(IMAGES_PATH), key=lambda x: int(x.split(".")[0]))

results = []

index = 1
ann_index = 1
for img in images:
    print("Processing file " + str(img))
    
    image = cv2.imread(os.path.join(IMAGES_PATH, img))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    nof, out_bboxes = detect_faces(image)
    
            
    for n in range(nof):
        data = {
            "image_id": index,
            "category_id": 1,
            "bbox": out_bboxes[n],
            "score": 1.0, 
        }
        results.append(data)
            

    
    index +=1
    
    if(index % 500 == 0):
        with open("part" + str(index) + ".json", "w") as f:
            json.dump(results, f)

with open(OUTPUT_PATH, "w") as f:
    json.dump(results, f)
```
